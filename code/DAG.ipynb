# Import librairies

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# DAG with white background

```python
# Create a Directed Acyclic Graph
dag = nx.DiGraph()

# Add nodes and edges to represent the relationships
dag.add_edges_from(
    [
        ("PredatorExperience", "PredatorBehavior"),
        ("PredatorBehavior", "PreyBehavior"),
        ("PredatorBehavior", "HuntingSuccess"),
        ("PreyBehavior", "HuntingSuccess"),
        ("PriorPreyBehavior", "PredatorBehavior"),
    ]
)

# Specify the positions of nodes to control arrow length
pos = {
    "PredatorExperience": (0.1, -0.005),
    "PredatorBehavior": (1, 0),
    "PreyBehavior": (2, 0),
    "HuntingSuccess": (1.5, -0.01),
    "PriorPreyBehavior": (0.1, 0.005),
}

# Define XKCD color RGB values
xkcd_colors = {
    "kelly green": "#02ab2e",
    "grass green": "#3f9b0b",
    "medium green": "#39ad48",
    "faded orange": "#f0944d",
    "melon": "#ff7855",
    "orange": "#f97306",
    "tangerine": "#ff9408",
    "blood orange": "#fe4b03",
}

# Draw the solid edges
nx.draw(
    dag, pos,
    with_labels = True,
    labels = {
        "PredatorExperience": "Cumulative\nhunting\nexperience",
        "PredatorBehavior": "Predator\nhunting\nbehavior",
        "PreyBehavior": "Prey\nbehavior",
        "HuntingSuccess": "Predator\nhunting\nsuccess",
        "PriorPreyBehavior": "Prior\nprey\nbehavior",
    },
    font_weight = "bold",
    font_size = 9,
    node_size = 5000,
    node_color = [
        xkcd_colors["medium green"],
        xkcd_colors["medium green"],
        xkcd_colors["orange"],
        xkcd_colors["medium green"],
        xkcd_colors["orange"]
    ],
    edgecolors = "black",
    arrowsize = 20,
    arrowstyle = "-|>",
    width = 3,
    style = "solid",
)

# Draw the curved solid arrow with curvature
start_pos_solid = pos["PreyBehavior"]
end_pos_solid = pos["PredatorBehavior"]

plt.annotate(
    "", xy = end_pos_solid, xytext = start_pos_solid,
    arrowprops = dict(
        arrowstyle = "-|>, head_width = 0.5, head_length = 0.7",
        linestyle = "solid", color = "black",
        connectionstyle = "arc3,rad = 0.7",
        shrinkA = 36,
        shrinkB = 37,
        linewidth = 3,
    ),
    annotation_clip = False
)

# Position of the curved dashed arrow
start_pos = pos["PredatorExperience"]
end_pos = pos["HuntingSuccess"]

# Draw dashed arrow with curvature using annotate
plt.annotate(
    "", xy = end_pos, xytext = start_pos,
    arrowprops = dict(
        arrowstyle = "-|>, head_width = 0.5, head_length = 0.7",
        linestyle = "dashed", color = "black",
        # Adjust the rad value as needed for curvature
        connectionstyle = "arc3,rad = 0.5",
        # Adjust the fractions to shorten the start and end of the arrow
        shrinkA = 37,
        shrinkB = 37,
        linewidth = 3,
    ),
    annotation_clip = False
)

# Add title
plt.title(
    "Predator-prey interactions\n mediated by predator experience",
    fontsize="xx-large")

# Save the plot
plt.savefig(
    "predator-prey_DAG.png",
    format = "png",
    bbox_inches = "tight"
)

# Display plot
plt.show()
```
